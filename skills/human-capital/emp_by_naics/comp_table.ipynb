```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("comp_1996_2020.csv")
print(len(df))
df = df.drop_duplicates(subset=['datadate', 'tic', 'naics'], keep='first')
print(len(df))
```

```python
df["naics_2"] = [str(ind)[:2] for ind in list(df.naics)]

df_96 = df[df.fyear == 1996].groupby(['naics_2']).sum()
df_96.reset_index(level=0, inplace=True)
df_96 = df_96[["naics_2", "emp"]]
df_a = pd.DataFrame({"naics_2":["31-33", "44-45", "48-49"], 
                     "emp":[float(df_96[df_96.naics_2 == "31"].emp) + float(df_96[df_96.naics_2 == "32"].emp) + float(df_96[df_96.naics_2 == "33"].emp),
                            float(df_96[df_96.naics_2 == "44"].emp) + float(df_96[df_96.naics_2 == "45"].emp),
                            float(df_96[df_96.naics_2 == "48"].emp) + float(df_96[df_96.naics_2 == "49"].emp)]})
df_96 = df_96.append(df_a, ignore_index=True)

df_06 = df[df.fyear == 2006].groupby(['naics_2']).sum()
df_06.reset_index(level=0, inplace=True)
df_06 = df_06[["naics_2", "emp"]]
df_a = pd.DataFrame({"naics_2":["31-33", "44-45", "48-49"], 
                     "emp":[float(df_06[df_06.naics_2 == "31"].emp) + float(df_06[df_06.naics_2 == "32"].emp) + float(df_06[df_06.naics_2 == "33"].emp),
                            float(df_06[df_06.naics_2 == "44"].emp) + float(df_06[df_06.naics_2 == "45"].emp),
                            float(df_06[df_06.naics_2 == "48"].emp) + float(df_06[df_06.naics_2 == "49"].emp)]})
df_06 = df_06.append(df_a, ignore_index=True)

df_16 = df[df.fyear == 2016].groupby(['naics_2']).sum()
df_16.reset_index(level=0, inplace=True)
df_16 = df_16[["naics_2", "emp"]]
df_a = pd.DataFrame({"naics_2":["31-33", "44-45", "48-49"], 
                     "emp":[float(df_16[df_16.naics_2 == "31"].emp) + float(df_16[df_16.naics_2 == "32"].emp) + float(df_16[df_16.naics_2 == "33"].emp),
                            float(df_16[df_16.naics_2 == "44"].emp) + float(df_16[df_16.naics_2 == "45"].emp),
                            float(df_16[df_16.naics_2 == "48"].emp) + float(df_16[df_16.naics_2 == "49"].emp)]})
df_16 = df_16.append(df_a, ignore_index=True)
```

```python
keys = ["11", "21", "22", "23", "31-33", "42", "44-45", "48-49", "51", "52", "53", "54", "56", "61", "62", "71", "72", "81"]
emp_1996 = []
emp_2006 = []
emp_2016 = []
for k in keys:
    emp_1996.append(float(df_96[df_96.naics_2 == k].emp))
    emp_2006.append(float(df_06[df_06.naics_2 == k].emp))
    emp_2016.append(float(df_16[df_16.naics_2 == k].emp))
```

```python
df_out = pd.DataFrame({"Industry":["Agriculture, Forestry, Fishing Hunting", 
                                   "Mining, Quarrying, Oil & Gas Extract", 
                                   "Utilities", 
                                   "Construction",
                                   "Manufacturing",
                                   "Wholesale Trade",
                                   "Retail Trade",
                                   "Transportation and Warehousing",
                                   "Information", 
                                   "Finance and Insurance",
                                   "Real Estate and Rental and Leasing",
                                   "Professional, Scientific, Tech. Services",
                                   "Administrative, Support, etc. Services",
                                   "Educational Services",
                                   "Health Care and Social Assistance",
                                   "Arts, Entertainment, and Recreation",
                                   "Accommodation and Food Services",
                                   "Other Services"],
                       "1996 Emp.":emp_1996,
                       "1996 Frac.": 100 * np.asarray(emp_1996) / sum(emp_1996),
                       "2006 Emp.":emp_2006,
                       "2006 Frac.": 100 * np.asarray(emp_2006) / sum(emp_2006),
                       "2016 Emp.":emp_2016,
                       "2016 Frac.": 100 * np.asarray(emp_2016) / sum(emp_2016)})
```
