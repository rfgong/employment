```python
import pandas as pd
from collections import OrderedDict 
import numpy as np
```

```python
df = pd.read_csv("qcew_2001_2019.csv")
```

```python
# Federal + State + Local + Private Employment
series_map = {}
for s in list(df["Series ID"]):
    k = s[-2:]
    if s[-3:-2] == "-":
        k = s[-5:]
    if k not in series_map:
        series_map[k] = list()
    series_map[k].append(s)

emp_2006 = OrderedDict()
emp_2016 = OrderedDict()
# skips 99
keys = list(series_map.keys())
keys.sort()
for i in range(len(keys)-1):
    val_2006 = 0
    val_2016 = 0
    for s in series_map[keys[i]]:
        try:
            val_2006 += int(df[df["Series ID"] == s]["Jan 2001"])
        except:
            print(df[df["Series ID"] == s]["Jan 2001"])
        try:
            val_2016 += int(df[df["Series ID"] == s]["Jan 2016"])
        except:
            print(df[df["Series ID"] == s]["Jan 2016"])
    emp_2006[keys[i]] = val_2006
    emp_2016[keys[i]] = val_2016
    
df_out = pd.DataFrame({"Industry":["Agriculture, Forestry, Fishing Hunting", 
                                   "Mining, Quarrying, Oil & Gas Extract", 
                                   "Utilities", 
                                   "Construction",
                                   "Manufacturing",
                                   "Wholesale Trade",
                                   "Retail Trade",
                                   "Transportation and Warehousing",
                                   "Information", 
                                   "Finance and Insurance",
                                   "Real Estate and Rental and Leasing",
                                   "Professional, Scientific, Tech. Services",
                                   "Administrative, Support, etc. Services",
                                   "Educational Services",
                                   "Health Care and Social Assistance",
                                   "Arts, Entertainment, and Recreation",
                                   "Accommodation and Food Services",
                                   "Other Services",
                                   "Public Administration"],
                      "2016 Emp.":list(emp_2016.values()),
                      "2016 Frac.": 100 * np.asarray(list(emp_2016.values())) / sum(emp_2016.values()),
                      "2006 Emp.":list(emp_2006.values()),
                      "2006 Frac.": 100 * np.asarray(list(emp_2006.values())) / sum(emp_2006.values())})
```
